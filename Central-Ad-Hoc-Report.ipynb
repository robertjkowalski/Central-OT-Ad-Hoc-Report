*Chart 3: Shows the total OT Hours usage by pay period in a tabular format* (This is supposed to be complimneted by an input tool 

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

def generate_usage_by_pay_period_chart(df):
    # Create a figure
    fig_usage_by_pp = go.Figure()

    # Add bars for Ops, Emg, and Total usage
    fig_usage_by_pp.add_trace(go.Bar(x=df['Pay_Periods'], y=np.round(df['Ops Hours Total'], 0), name='Ops Hours Total', marker_color='red'))
    fig_usage_by_pp.add_trace(go.Bar(x=df['Pay_Periods'], y=np.round(df['Emg Hours Total'], 0), name='Emg HoursTotal', marker_color='blue'))
    fig_usage_by_pp.add_trace(go.Bar(
        x=df['Pay_Periods'],
        y=(df['Ops Hours Total'] + df['Emg Hours Total']).cumsum(),  # Cumulative sum of Ops Total + Emg Total
        name='Total',
        marker_color='green'))

    # Update layout
    fig_usage_by_pp.update_layout(barmode='group', xaxis_title='Pay Period', yaxis_title='Total OT Hours Used', title='Usage by Pay Period')

    # Display the figure
    fig_usage_by_pp.show()

    # Create a new figure for the table
    fig_usage_by_pp_table = go.Figure()

    # Create the table trace
    fig_usage_by_pp_table.add_trace(go.Table(
        header=dict(values=['Pay Periods', 'Ops Hours Total', 'Emg Hours Total', 'Total'],
                    fill_color='paleturquoise',
                    align='left'),
        cells=dict(values=[df['Pay_Periods'],
                            np.round(df['Ops Hours Total'], 0),
                            np.round(df['Emg Hours Total'], 0),
                            np.round((df['Ops Hours Total'] + df['Emg Hours Total']).cumsum(), 0)],
                   fill_color='lavender',
                   align='left'))
    )

    # Update layout
    fig_usage_by_pp_table.update_layout(title='Usage by Pay Period (Tabular Format)')

    # Display the table
    fig_usage_by_pp_table.show()

# Read data from the Excel file
df = pd.read_excel("C:/Users/robert.kowalski/OneDrive - City of Dallas-5516124-G Tenant/Jupyter Notebook/Central-Py.xlsx")

# Call the function to generate the bar chart and table
generate_usage_by_pay_period_chart(df)
```
